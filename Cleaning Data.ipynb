```python
# import module
import pandas as pd
  
# read CSV file
results = pd.read_csv('games.csv')
  
results = results.drop(columns=['GAME_STATUS_TEXT','TEAM_ID_home','TEAM_ID_away', 'HOME_TEAM_WINS'])
results = results.dropna()
results.to_csv('./games_updated.csv', index=False)
```

```python
game_details = pd.read_csv('games_details.csv')
game_details = game_details.drop(columns=['TEAM_ABBREVIATION','TEAM_CITY','NICKNAME', 'START_POSITION','COMMENT', 'PLUS_MINUS'])
game_details = game_details.dropna()

game_details.to_csv('./game_details_updated.csv', index=False)
  
```

```python
players = pd.merge(game_details, results, on='GAME_ID', how='left')
players = players[['PLAYER_NAME','TEAM_ID','PLAYER_ID', 'SEASON']]
players = players.drop_duplicates()

players[players['PLAYER_NAME'] == 'LeBron James']
```

```python
salaries = pd.read_csv('salaries_1985to2018.csv')
players2 = pd.read_csv('players2.csv')
players2 = players2.rename(columns={'_id': 'player_id'})
players2
merge = pd.merge(salaries, players2, on='player_id', how='left')
merge = merge[['season_end', 'salary', 'name', 'team']]
merge[merge['name'] == 'LeBron James']
merge = merge.rename(columns={'name': 'PLAYER_NAME', 'season_end': 'SEASON', 'salary': 'SALARY'})
merge['TEAM_ID'] = merge['team'].apply(lambda x: getTeamID(x))
merge = merge.drop(columns=['team'])

players3 = pd.merge(players, merge, on=['TEAM_ID', 'SEASON', 'PLAYER_NAME'], how='left')
players3[players3['PLAYER_NAME'] == 'LeBron James']

players3.to_csv('./players_updated.csv', index=False)
```
